```python
import os, gensim, nltk, re, string
import pandas as pd
from nltk.tokenize import TreebankWordTokenizer
from gensim.test.utils import common_texts
from gensim.models import Word2Vec

nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
corpus_sents = []
corpus_name = "augmented_data/Dataset_aug_complex_6513_original.csv"
model_path = ""
model_name = "word2vec.model"
tokenizer = TreebankWordTokenizer()

corpus_df = pd.read_csv(corpus_name, sep=',')

for t in corpus_df['text'] :
    corpus_sents.append(tokenizer.tokenize(t.lower()))

model = Word2Vec(sentences=corpus_sents, vector_size=100, trim_rule= lambda x,y,z: gensim.utils.RULE_KEEP)
model.save(model_path+model_name)
```

```python
stop_words = set(stopwords.words('english'))
re_punkt = re.compile("[" + string.punctuation + "]+")

train_df = corpus_df.sample(5000)
for t in train_df['text'] :
    tokenized = tokenizer.tokenize(t)
    tokens_no_punkt_stopwords = [i.lower() for i in tokenized if (i.lower() not in stop_words) and (not re_punkt.fullmatch(i))]
    for token in tokens_no_punkt_stopwords :
        print(model.wv[token])
```
